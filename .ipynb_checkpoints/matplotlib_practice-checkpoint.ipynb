```python
#import dependencies
import matplotlib.pyplot as plt


# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#create the plot MATLAB way
#plt.plot(x_axis, y_axis)
#plt.show()

#Object oriented way, 
    #create figure
#fig, ax = plt.subplots()
#ax.plot(x_axis,y_axis)

#fig, ax = plt.subplots() same as
#fig = plt.figure()
#ax = fig.add_subplot()
#ax.plot(x_axis,y_axis)

#quickest way to creat line graph
ax = plt.axes()
ax.plot(x_axis,y_axis)
```

```python
#line chart with formatting Object Oriented way
fig, ax = plt.subplots()
ax.plot(x_axis,y_axis, marker = 'd', color = 'green', linewidth = 2, label = 'Boston')
ax.grid()
ax.set_title('PyBer Fare by Month')
ax.set_xlabel('Date')
ax.set_ylabel('Fare(s)')
ax.set_ylim(0,45)
ax.legend()



```

```python
#MATLAB WAY BAR
# Create the plot.
plt.barh(x_axis, y_axis, color="magenta", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Fare($)")
plt.ylabel("Date")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()

plt.gca().invert_yaxis()
```

```python
#Object Oriented Bar
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color = 'cyan', label='Chicago')
ax.legend()
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare")
ax.set_ylabel("Date")
ax = plt.gca()
ax.invert_yaxis()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]


plt.plot(y_axis, x_axis, 'o', c = 'red', label = "Chicago")
plt.title("PyBer Fare by Month")
plt.gca().invert_yaxis()
plt.xlim(0,50)
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s = [i*5 for i in y_axis], c = 'skyblue', linewidth = 2, alpha = .8, edgecolors = 'black')

```

```python
#pie charts
#MATLAB way
explode = (0,0,0,0,0,0,.2,0,0,0,0,0)
plt.subplots(figsize =(8,8))
plt.pie(y_axis, labels = x_axis, autopct = "%.1f%%", explode = explode)
plt.show()
```

```python
#pie chart object oriented
explode = (0,0,.2,0,0,0,.2,0,0,0,0,0)
fig, ax = plt.subplots(figsize = (8,8))
ax.pie(y_axis, labels = x_axis, autopct = "%.1f%%", explode = explode, shadow = True, startangle=80)


```
